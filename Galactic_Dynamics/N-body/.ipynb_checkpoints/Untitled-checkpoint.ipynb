# Two-body

## Gravity

$$
-\frac{G\mu}{|\mathbf x|^3}\mathbf x=\ddot{\mathbf x}
$$

## Dimensionless formalism of gravity


```python
GM = 1


def Fg(x, t):
    r = ((x**2).sum())**(1 / 2)
    return -1/r**3 * x * GM


def E(X):
    x, v = X[0], X[1]
    r = ((x**2).sum())**(1 / 2)
    return -GM / r + (v**2).sum() / 2.
```

```python
dT = 1e-3
Tmax = 100
```

```python
t = 0
X = [np.array([1, 0, 0]), np.array([0.05, 0.5, 0])]
x_his, v_his = [X[0]], [X[1]]
E_his = []
t_his = []
while t < Tmax:
    X, t = Runge_Kutta(X, t, Fg, dT)
    x_his.append(X[0])
    v_his.append(X[1])
    E_his.append(E(X))
    t_his.append(t)
x_his = np.array(x_his)
plt.plot(x_his[:, 0], x_his[:, 1])
plt.axis('equal')
plt.show()
plt.plot(t_his, E_his-E_his[0])
plt.show()
```

```python
t = 0
X = [np.array([1, 0, 0]), np.array([0.05, 0.5, 0])]
x_his, v_his = [X[0]], [X[1]]
E_his = []
t_his = []
while t < Tmax:
    X, t = LeapFrog(X, t, Fg, dT)
    x_his.append(X[0])
    v_his.append(X[1])
    E_his.append(E(X))
    t_his.append(t)
x_his = np.array(x_his)
plt.plot(x_his[:, 0], x_his[:, 1])
plt.axis('equal')
plt.show()
plt.plot(t_his, E_his-E_his[0])
plt.show()
```
